```python
mass=df["mass_grams"]
fall_status=df["fall_status"]
top5_havy=df.sort_values(by="mass_grams",ascending=False).iloc[0:5]
print(top5_havy)
filtered=df[
    (df["mass_grams"]>50000)&
    ((df["region"].str.contains("Antarctica"))|(df["region"].str.contains("Sahara")))
]
print(filtered)
```
