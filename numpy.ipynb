2. Imagine you have two 3x3 matrices represented as lists of lists. How would you write the standard Python code to add them together, number by number?

```python
import numpy as py
x=py.array([[1,2,3],[2,3,4],[4,5,6]])
y=py.array([[3,4,5],[1,5,6],[2,3,4]])
print(py.stack((x,y),axis=1))
```
