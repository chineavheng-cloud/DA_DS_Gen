The Goal: Create a base Media class with a play() method. Then, create subclasses like Song, Movie, and Podcast that inherit from Media and override the play() method to output unique processing messages (e.g., "Playing audio...", "Playing video at 1080p...", "Playing podcast..."). Finally, write a standalone function play_all(media_list) that loops through a list containing different media objects and triggers the play() method on each one, demonstrating how a single function handles different object types interchangeably. 
Concepts Tested: Polymorphism, Method Overriding, Inheritance, Duck Typing.

```python
class Media:
    def play(self):
        pass
class Song(Media):
    def play(self):
        return "Playing audio"
class Movie(Media):
    def play(self):
        return "Playing video at 1080p"
class Podcast(Media):
    def play(self):
        return "Playing podcast"
def play_media(media_list):
    for i in media_list:
        print(i.play())
m=[Song(),Movie(),Podcast()]
play_media(m)
```

The Goal: Create a DigitalWallet class with private attributes for the account __balance and a __pin_code. Create methods like deposit(), withdraw(), and get_balance() that interact with these attributes securely. Ensure that withdraw() and get_balance() require the user to pass the correct PIN to work, and that the __balance cannot be directly accessed, modified, or set to a negative value from outside the class. 
Concepts Tested: Encapsulation, private attributes (name mangling), data hiding, getter/setter methods, and access control.

```python
class Wallet:
    def __init(self,balance,pin_code,initial_balance=0):
        self.__balance=initial_balance
        self.__pin_code=pin_code
    def deposit(self,amount):
        if amount<=0:
            print("Deposit amount must be pisitive.")
            return
        self.__balance+=amount
        print(f"${amount} deposited.")
    def withdraw(self,amount,pin_code):
        if pin_code!=self.__pin_code:
            print("Incorrect PIN code.")
        if amount<=0:
            print("Withdrawal amount must be positive.")
            return
        if amount>self.__balance:
            print("Amount exceeds balance.")
            return
        self.__balance-=amount
        print(f"${amount} withdrawn.")
    def get_balance(self,pin_code):
        if pin_code!=self.__pin_code:
            print("Incorrect PIN.")
            return
        print(f"Current balance: ${self.__balance}")
w=Wallet(1000,"4521")
w.deposit(500)
w.withdraw(200,"4521")
w.get_balance("4521")
```
