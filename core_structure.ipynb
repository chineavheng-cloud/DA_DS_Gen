Memory-Based To-Do List:
The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
Concepts Tested: List methods (.append(), .pop()), indexing, iterating through lists with for loops.

```python
task=[]
n=int(input("How many task to do today? "))
for i in range(n):
    to_do=input(f"Enter task {i+1}: ")
    task.append(to_do)
print("\nTask to do today: ")
for i in task:
    print("- "+i)
remove=input("\nEnter the task to remove: ")
if remove in task:
    task.remove(remove)
    print("Task removed!")
else:
    print("Task not found!")
print("\nUpdated task list:")
for i in task:
    print("- "+i)
```

Interactive Flashcard / Quiz App:
The Goal: Create a dictionary where the keys are questions (e.g., countries) and values are answers (e.g., capitals). Loop through the dictionary, ask the user the question, check their answer, and calculate a final score.
Concepts Tested: Dictionaries, key-value pairs, iteration, score tracking.

```python
country={
    "cambodia":"Phnom Penh",
    "japan":"Tokyo",
    "northkorea":"Pyongyang",
    "southkorea":"Seoul",
    "nepal":"Kathmandu"
}
score=0
print("Let's answer 5 question")
for key,value in country.items():
    ask=input(f"What is the capital of {key}? ")
    if ask.lower()==value.lower():
        print(f"You're correct!")
        score+=20
    else:
        print(f"The capital of {key} is {value}.")
print(f"\nFinal score: {score}")
```

The Shopping Cart Analyzer
The Goal: Ask the user to input the prices of items they are adding to their cart, one by one. Store these prices in a Python List. The loop stops when the user types 'done'. Then, calculate and print a receipt showing the total cost, the most expensive item in the cart, and the average item price.
Concepts Tested: while loops for input collection, list methods (.append()), built-in list functions (sum(), max(), len()), and data type casting (handling the mix of string commands and float prices).

```python
lists=[]
while True:
    ch=input("Enter iteam price or type 'done': ")
    if ch.lower()=="done":
        break
    try:
        price=float(ch)
        lists.append(price)
    except ValueError:
        print("Invalid input!Pls enter number or 'done'.")
if len(lists)==0:
    print("\nYour lists is empty")
else:
    total=sum(lists)
    maxValue=max(lists)
    avg=total/len(lists)
    print("\n---Receipt---")
    print(f"Total cost: ${total:.2f}\nMost expensive item: ${maxValue:.2f}\nAverage price: ${avg:.2f}")
```

The Coffee Shop Inventory Manager
The Goal: Create a dictionary pre-filled with cafe supplies and their quantities (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks the user if they want to "sell" or "restock" an item. Update the dictionary values based on their input and print the current inventory levels. Prevent the user from selling more items than they have in stock!
 Concepts Tested: Accessing and updating dictionary values, checking if a key exists using the in operator, combining if/elif logic with dictionary lookups, and managing state.

```python
pre_filled={
    "coffee beans":50,
    "cup":100,
    "syrup":15
}
while True:
    print("\nCurrent Inventory:")
    for k,v in pre_filled.items():
        print(f"{k}: {v}")
    ask=input("What you want to do?(sell/restock/exit): ").lower()
    if ask=="exit":
        break
    if ask=="sell":
        ch=input("Enter item name: ").lower()
        if ch not in pre_filled:
            print("No item found!")
            continue
        try:
            qty=int(input("Enter quantity: "))
        except ValueError:
            print("Pls enter a valid number!")
            continue
        if qty>pre_filled[ch]:
            print("Not enough in stock!")
        else:
            pre_filled[ch]-=qty
            print(f"Sold {qty} {ch}")
    elif ask=="restock":
        pre_filled[ch]+=qty
        print(f"Restocked {qty} {ch}")
    else:
        print("Invalid option! Try again. ")
```
