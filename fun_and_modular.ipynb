Modular Unit Converter:
The Goal: A program that converts temperatures (Celsius to Fahrenheit), distances (Miles to Kilometers), and weights(Grams to Kilograms).
The Catch: Every single conversion must be its own independent function (e.g., def c_to_f(celsius):). There should be a main menu function that calls the others based on user choice.
Concepts Tested: Defining functions, passing arguments, return statements, variable scope.

```python
def c_to_f(celsius):
    f=(celsius* 9/5)+32
    return f
def m_to_km(m):
    km=m*1.609
    return km
def g_to_kg(g):
    kg=g/1000
    return kg
while True:
    print("Which one you should do?")
    op=input("Coverts: \n\t(Celsius to Fahrenheit)c, \n\t(Miles to kilometers)m, \n\t(Grams to kilograms)g, or exit:  ").lower()
    if op=="exit":
        break
    if op=="c":
        celsius=float(input("How many celsius? "))
        print("Farenheit= ",c_to_f(celsius))
    elif op=="m":
        m=float(input("How many miles? "))
        print("Kilometers= ",m_to_km(m))
    elif op=="g":
        g=float(input("How many gram? "))
        print("Kilograms= ",g_to_kg(g))
    else:
        print("Invalid option! pls try again!")
```

Password Generator:
The Goal: A function that takes an integer (length) and boolean flags (use_numbers, use_symbols) and returns a randomized, secure password.
Concepts Tested: Default arguments, keyword arguments, combining lists/strings, returning data.

```python
import random
import string
def gen_pass(length=8,use_numbers=True,use_symbols=True):
    ch=list(string.ascii_letters)
    if use_numbers:
        ch+=list(string.digits)
    if use_symbols:
        ch+=list("!@#$%^&*()_=+{}[]|;:,.<>?/")
    if length>len(ch):
        password="".join(random.choice(ch,k=length))
    else:
        password="".join(random.sample(ch,k=length))
    return password
ask=input("Which option you choice (normal genrate, no symbol,no numbers)(g,s,n): ")
if ask=='g':
    print("Password : ",gen_pass(12))
elif ask=='s':
    print("Password : ",gen_pass(16,use_symbols=False))
elif ask=='n':
    print("Password : ",gen_pass(10,use_numbers=False))
```

The Text Analyzer Suite
The Goal: Ask the user to input a long sentence or paragraph. Build a program that outputs statistics about their text.
The Catch: You must use separate functions for every analysis: count_words(text), count_vowels(text), and find_longest_word(text). Each function must take the text as an argument and return the result to be printed by the main program.
Concepts Tested: String manipulation, passing strings as arguments, returning specific values (integers and strings) from functions, and keeping logic strictly separated.

```python
def word(text):
    words=text.split()
    return len(words)
def vowel(text):
    vowel="aeiouAEIOU"
    count=sum(1 for char in text if char in vowel)
    return count
def longest(text):
    word=text.split()
    if not word:
        return ""
    long=max(word,key=len)
    return long
text=input("Enter a long sentences or a paragraph: \n\t")
print("\n---Text Analysis---")
print(f"""Total words: {word(text)}
Total Vowels: {vowel(text)}
Longest Words: {longest(text)}""")
```

The Virtual ATM (Banking Simulator)
The Goal: Simulate a bank account where a user can deposit, withdraw, or check their balance.
The Catch: Create specific functions like deposit(balance, amount) and withdraw(balance, amount). These functions must return the newly calculated balance to overwrite the old balance in the main program loop. The withdraw function must also include logic to prevent overdrafting (returning an error message and the unmodified balance if they ask for too much).
Concepts Tested: Passing variables as arguments, updating state using return statements, handling edge cases within functions, and understanding variable scope (realizing that a variable modified inside a function doesn't change outside unless returned).

```python
def deposit(balance,amount):
    if amount<=0:
        print("Deposit must be greater than 0 ")
        return balance
    return balance+amount
def withdraw(balance,amount):
    if amount<=0:
        print("Withdrawl must be greater than 0 ")
        return balance
    if amount>balance:
        print("Not enough money in balance ")
        return balance
    return balance-amount
balance=0.0
while True:
    print("\n---Virtual ATM---\n1.Deposit\n2.Withdraw\n3.Check balance\n4.Exit")
    ch=input("Choose an option (1-4): ")
    if ch=="1":
        amount=float(input("Enter amount to deposit: "))
        balance=deposit(balance,amount)
        print("Deposit successful")
    elif ch=="2":
        amount=float(input("Enter amount to withdraw: "))
        balance=withdraw(balance,amount)
        print("Withdraw successful")
    elif ch=="3":
        print(f"Your current balance is {balance}")
    elif ch=="4":
        break
    else:
        print("Invalid option")
```
