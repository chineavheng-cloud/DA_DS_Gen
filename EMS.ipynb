The Goal: Create a base Employee class with the attributes name and base_salary. Include a method called calculate_pay() that simply returns the base salary.
Next, create two child classes that inherit from Employee: Manager and Developer.
For the Manager class: Use the super() function in the __init__ method to inherit the base attributes, but add a new, specific attribute called bonus. Override the inherited calculate_pay() method so that it returns the base_salary plus the bonus.
For the Developer class: Use super() to inherit the base attributes and add a new attribute called programming_language. Create a brand new, unique method called write_code() that prints a message stating the developer's name and the language they are coding in (e.g., "Alice is writing code in Python").
Concepts Tested: Parent and Child classes, the super() function, Method Overriding, and extending child classes with new, unique methods and attributes.

```python
class emp:
    def __init__(self,name,base_salary):
        self.name=name
        self.base_salary=base_salary
    def calculate_pay(self):
        return self.base_salary
class manager(emp):
    def __init__(self,name,base_salary,bonus):
        super().__init__(name,base_salary)
        self.bonus=bonus
    def calculate_pay(self):
        return super().calculate_pay()+self.bonus
class developer(emp):
    def __init__(self,name,base_salary,programming_language):
        super().__init__(name,base_salary)
        self.programming_language=programming_language
    def write_code(self):
        print(f"{self.name} is writing code in {self.programming_language}")

e=emp("Seav eng",1500)
m=manager("Heng",3500,500)
d=developer("Samon",2000,"Python")
print("Employee pay: ",e.calculate_pay())
print("Manager pay: ",m.calculate_pay())
print("Developer pay: ",d.calculate_pay())
d.write_code()
```
