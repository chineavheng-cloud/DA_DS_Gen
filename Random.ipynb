Number Guessing Game:
The Goal: The program randomly selects a number between 1 and 100. The user guesses the number, and the program replies with "Too high!", "Too low!", or "You got it!". The game loops until the user guesses correctly.
Concepts Tested: while loops, if/elif/else statements, comparison operators, importing the basic random module.

```python
import random
num=random.randint(1,100)
def guess():
    ans=int(input("Guessing the number from 1-100 "))
    return ans
while True:
    ans=guess()
    if(ans<num):
        print("Too low!")
        guess()
    elif(ans>num):
        print("Too high!")
        guess()
    else:
        print("You got it")
        break
```

Rock, Paper, Scissors:
The Goal: A continuous loop where the user plays Rock, Paper, Scissors against the computer. Keep track of the score until the user types "quit".
Concepts Tested: Infinite loops with break conditions, complex if/elif logic, string normalization (e.g., using .lower()).

```python
import random
ch=["rock","paper","scissors"]
point=0
while True:
    
    user=input("Enter rock,paper,scissors, e for exit: ").lower()
    ai=random.choice(ch)
    if user=='e':
        break
    if user==ai:
        print("Tie!")
    elif(user=="rock" and ai=="scissors")or\
        (user=="paper" and ai=="rock")or\
        (user=="scissors"and ai=="paper"):
        print("You got 5 point")
        point+=5
    else:
        print("You lose!")
print(f"You got {point} point for this game.")
```

The Virtual Barista (Simple Point-of-Sale)
The Goal: Display a text menu of three coffee shop drinks with different prices. Create a continuous loop that asks the user what they want to order. Add the price of their chosen drink to a running total. If they type an invalid drink, tell them it's not on the menu. The loop should continue until the user types "checkout", at which point the program prints their final total.
Concepts Tested: Infinite while loops, break conditions, accumulating totals (+= operator), complex if/elif/else logic, and string normalization (using .lower() to handle if the user types "Checkout" or "CHECKOUT").

```python
drinks=["matcha","ice latte","green tea"]
prices={"matcha":4.05,"ice latte":2.5,"green tea":1.8}
total=0.0
while True:
    print("Menu: ",", ".join(drinks))
    order=input("What you want to order or no need(checkout)? ").lower().replace(" ","")
    if order=="checkout":
        break
    mat=False
    for i in drinks:
        if order==i.replace(" ",""):
            print(f"You ordered {order}")
            total+=prices[i]
            mat=True
    if not mat:
        print("\nInvalid drinks!")
print(f"Total price: ${total:.2f}")
```

If the user gets it right, they get a point, and the program asks another question.
