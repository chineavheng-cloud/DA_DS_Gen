```python
totalBill=float(input("Total bill: "))
ch=input("Do you want to tip?(y,n): ")
if(ch=='y'):
    tip=float(input("Enter your tip: "))
    totalBill+=tip
print(f"{totalBill:.2f}")
```

```python
def round(km,kpl):
    avg=(km/10)*kpl
    price=avg*1.35
    return price
kpl=1.5
km=float(input("How far the planned road trip? "))
print(f"Total price for a round: ${round(km,kpl):.2f}",)
```
