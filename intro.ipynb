Restaurant Tip Calculator:
The Goal: Ask the user for the total bill amount and the percentage they want to tip. Calculate the tip amount and the grand total, rounding the output to two decimal places.
Concepts Tested: Data type casting (converting string inputs to floats), basic arithmetic operators.

```python
totalBill=float(input("Total bill: "))
ch=input("Do you want to tip?(y,n): ")
if(ch=='y'):
    tip=float(input("Enter your tip: "))
    totalBill+=tip
print(f"{totalBill:.2f}")
```

The Road Trip Cost Estimator
The Goal: Ask the user for the total distance of their planned road trip (in km), their car's average KPL (kilometers per liter), and the current price of fuel per liter. Calculate how many liter they will need and the total estimated cost of the trip, outputting a cleanly formatted summary with the total rounded to two decimal places.
Concepts Tested: input(), print(), data type casting (converting string inputs to floats), basic arithmetic operators (division and multiplication), f-strings, and the round() function.

```python
def round(km,kpl):
    avg=(km/10)*kpl
    price=avg*1.35
    return price
kpl=1.5
km=float(input("How far the planned road trip? "))
print(f"Total price for a round: ${round(km,kpl):.2f}",)
```
